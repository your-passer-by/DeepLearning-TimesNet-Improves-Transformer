```python
def plot_accuracy_vs_threshold(thresholds, accuraciesp):
    plt.plot(thresholds, accuraciesp[0], marker='o', color='b', label='Timesnet Transformer Accuracy vs Threshold')
    plt.plot(thresholds, accuraciesp[1], marker='o', color='g', label='transformer Accuracy vs Threshold')
    plt.title('Accuracy vs Threshold')
    plt.xlabel('Threshold')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()
```
